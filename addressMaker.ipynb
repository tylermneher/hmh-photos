```python
from openpyxl import load_workbook
import requests
import pandas as pd
```

```python
workbook = load_workbook(filename="photos_info.xlsx")
sheet = workbook.active

print(f'workbook.sheetnames = {workbook.sheetnames}')
print(f'sheet = {sheet}')
```

```python
ids = []
lats = []
lons = []
cols_orig = []

def create_url(lat, lon):
    url = f'https://nominatim.openstreetmap.org/reverse?format=json&zoom=18&namedetails=1&addressdetails=1&extratags=1&email=dev@neherdata.com&lat={str(lat)}&lon={str(lon)}'
    return url

def dicts_from_urls(url):
    r = requests.get(url)
    j = r.json()
    return j

# col names
for value in sheet.iter_rows(min_row=1, max_row=1, values_only=True):
    for i in value:
        cols_orig.append(str(i))

# lists of lats and lons and ids
for value in sheet.iter_rows(min_row=2, values_only=False):
    id = str(value[0].value)
    ids.append(id)
    lat = str(value[3].value)
    lats.append(lat)
    lon = str(value[4].value)
    lons.append(lon)
```

```python
# make df and populate urls column
df = pd.read_excel("photos_info.xlsx")
df.dropna(subset=['latitude', 'longitude'], inplace=True)

df['url'] = df.apply(lambda row: create_url(row['latitude'], row['longitude']), axis=1)

df
```
